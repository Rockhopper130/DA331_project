```python
def parse(x):
    # otpt = np.array2string(x, separator=' ')
    otpt = x
    otpt = otpt.replace("\n", "")
    return otpt

df["embeddings_distilbert"] = df["embeddings_distilbert"].apply(lambda x: parse(x))
```
